## ✅ Task 1: Replace Hardcoded Prompts (10%)

**Goal:** Convert a hardcoded f-string function into a reusable `PromptTemplate`.

**Given Code (Bad Practice ❌):**
```python
def explain_topic(topic):
    return f"Explain {topic} in simple terms for beginners"
```

**Requirements:** No f-strings · Use `PromptTemplate` · Must be reusable


````python
# ── Task 1: Replace Hardcoded Prompts ─────────────────────────────────────────

# ❌ Old approach — hardcoded f-string (DO NOT USE)
# def explain_topic(topic):
#     return f"Explain {topic} in simple terms for beginners"

# ✅ New approach — LangChain PromptTemplate
explain_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

def explain_topic(topic: str) -> str:
    """Generates a beginner-friendly explanation prompt for a given topic."""
    return explain_template.format(topic=topic)

# ── Test Task 1 ────────────────────────────────────────────────────────────────
print("Task 1 — Reusable PromptTemplate Output:")
print("-" * 45)
print(explain_topic("Machine Learning"))
print(explain_topic("Neural Networks"))
print(explain_topic("Data Science"))
````

## ✅ Task 2: Multi-Input Prompt System (15%)

**Template:** `Explain {topic} for {audience} in a {tone} tone`

**Test Cases:**
| topic | audience | tone |
|---|---|---|
| AI | beginners | friendly |
| Python | kids | fun |
| Deep Learning | engineers | technical |


````python
# ── Task 2: Multi-Input Prompt System ─────────────────────────────────────────

multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

def explain_for_audience(topic: str, audience: str, tone: str) -> str:
    """Generates a customised explanation prompt based on topic, audience, and tone."""
    return multi_input_template.format(topic=topic, audience=audience, tone=tone)

# ── Test Cases ─────────────────────────────────────────────────────────────────
test_cases = [
    ("AI",            "beginners", "friendly"),
    ("Python",        "kids",      "fun"),
    ("Deep Learning", "engineers", "technical"),
]

print("Task 2 — Multi-Input Prompt System Output:")
print("-" * 50)
for topic, audience, tone in test_cases:
    prompt = explain_for_audience(topic, audience, tone)
    print(f"  → {prompt}")
````

## ✅ Task 3: Prompt Variations Engine (15%)

**Three style templates for the same topic:**
- **Teaching** → Step-by-step explanation
- **Interview** → 3 interview questions
- **Storytelling** → Topic as a story


````python
# ── Task 3: Prompt Variations Engine ─────────────────────────────────────────

# Three distinct PromptTemplates — same input, different style
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

storytelling_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

# Bundle all variations into a dictionary for easy access
prompt_variations = {
    "teaching":     teaching_template,
    "interview":    interview_template,
    "storytelling": storytelling_template,
}

def generate_variation(topic: str, style: str) -> str:
    """Returns a styled prompt for the given topic and style variation."""
    template = prompt_variations.get(style)
    if not template:
        raise ValueError(f"Unknown style '{style}'. Choose from: {list(prompt_variations.keys())}")
    return template.format(topic=topic)

# ── Run for topic = Machine Learning ──────────────────────────────────────────
topic = "Machine Learning"
print(f"Task 3 — Prompt Variations for topic: '{topic}'")
print("-" * 55)
for style in ["teaching", "interview", "storytelling"]:
    print(f"  [{style.upper()}]  →  {generate_variation(topic, style)}")
````

## ✅ Task 4: ChatPromptTemplate System (15%)

**Build a chat-based prompt with System + User roles.**

| Role | Description |
|---|---|
| System | Defines the AI's behaviour |
| User | Dynamic topic input |

Roles supported: `teacher`, `interviewer`, `motivator`


````python
# ── Task 4: ChatPromptTemplate System ────────────────────────────────────────

# System behaviour definitions per role
role_behaviours = {
    "teacher":     "You are an expert teacher who explains concepts clearly and patiently.",
    "interviewer": "You are a professional interviewer who asks sharp, insightful questions.",
    "motivator":   "You are an enthusiastic motivator who inspires people to learn new things.",
}

def build_chat_prompt(role: str, topic: str) -> list:
    """
    Builds a ChatPromptTemplate for the given role and topic.
    Returns a list of formatted chat messages.
    """
    if role not in role_behaviours:
        raise ValueError(f"Unknown role '{role}'. Choose from: {list(role_behaviours.keys())}")

    chat_template = ChatPromptTemplate.from_messages([
        SystemMessagePromptTemplate.from_template(
            "{role_description}"
        ),
        HumanMessagePromptTemplate.from_template(
            "Help me understand {topic}."
        )
    ])

    messages = chat_template.format_messages(
        role_description=role_behaviours[role],
        topic=topic
    )
    return messages

# ── Test with role='teacher', topic='Neural Networks' ─────────────────────────
print("Task 4 — ChatPromptTemplate System Output:")
print("-" * 55)
for role in ["teacher", "interviewer", "motivator"]:
    messages = build_chat_prompt(role=role, topic="Neural Networks")
    print(f"\n  Role: {role.upper()}")
    for msg in messages:
        msg_type = type(msg).__name__.replace("Message", "")
        print(f"    [{msg_type}]: {msg.content}")
````

## ✅ Task 5: Input Validation Layer (10%)

**Rules:**
- `audience` must be in `[beginner, intermediate, expert]`
- `tone` must be in `[formal, casual, fun]`
- If invalid → raise error OR assign default values


````python
# ── Task 5: Input Validation Layer ──────────────────────────────────────────

VALID_AUDIENCES = ["beginner", "intermediate", "expert"]
VALID_TONES     = ["formal", "casual", "fun"]

DEFAULT_AUDIENCE = "beginner"
DEFAULT_TONE     = "casual"

def validate_inputs(audience: str, tone: str, use_defaults: bool = True) -> tuple:
    """
    Validates audience and tone inputs.
    
    Args:
        audience      : Target audience level
        tone          : Communication tone
        use_defaults  : If True, assign defaults on invalid input.
                        If False, raise a ValueError.
    
    Returns:
        Tuple of (validated_audience, validated_tone)
    """
    validated_audience = audience.lower().strip()
    validated_tone     = tone.lower().strip()

    # Validate audience
    if validated_audience not in VALID_AUDIENCES:
        if use_defaults:
            print(f"  ⚠️  Invalid audience '{audience}' → defaulting to '{DEFAULT_AUDIENCE}'")
            validated_audience = DEFAULT_AUDIENCE
        else:
            raise ValueError(
                f"Invalid audience '{audience}'. Must be one of: {VALID_AUDIENCES}"
            )

    # Validate tone
    if validated_tone not in VALID_TONES:
        if use_defaults:
            print(f"  ⚠️  Invalid tone '{tone}' → defaulting to '{DEFAULT_TONE}'")
            validated_tone = DEFAULT_TONE
        else:
            raise ValueError(
                f"Invalid tone '{tone}'. Must be one of: {VALID_TONES}"
            )

    return validated_audience, validated_tone

# ── Test Cases ─────────────────────────────────────────────────────────────────
print("Task 5 — Input Validation Layer:")
print("-" * 50)

print("\n  [Valid inputs]")
aud, tone = validate_inputs("beginner", "formal")
print(f"    audience='{aud}', tone='{tone}'")

print("\n  [Invalid audience — use_defaults=True]")
aud, tone = validate_inputs("ninja", "fun")
print(f"    audience='{aud}', tone='{tone}'")

print("\n  [Invalid tone — use_defaults=True]")
aud, tone = validate_inputs("expert", "yelling")
print(f"    audience='{aud}', tone='{tone}'")

print("\n  [Invalid inputs — use_defaults=False → raises error]")
try:
    validate_inputs("unknown", "weird", use_defaults=False)
except ValueError as e:
    print(f"    ❌ ValueError caught: {e}")
````

## ✅ Task 6: Prompt Generator App (15%)

**Function:** `generate_prompt(topic, audience, tone, style)`

Combines validation + template selection into one clean pipeline.

**Styles:** `teaching`, `interview`, `storytelling`


````python
# ── Task 6: Prompt Generator App ────────────────────────────────────────────

# Style templates with full context (topic + audience + tone)
style_templates = {
    "teaching": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} teaching style, step by step"
    ),
    "interview": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Generate 3 {tone} interview questions on {topic} suitable for a {audience}"
    ),
    "storytelling": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} storytelling style"
    ),
}

def generate_prompt(topic: str, audience: str, tone: str, style: str) -> str:
    """
    Full prompt generation pipeline:
    User Input → Validation → Template Selection → Dynamic Prompt Output

    Args:
        topic    : Subject to explain
        audience : Target audience (beginner / intermediate / expert)
        tone     : Communication tone  (formal / casual / fun)
        style    : Prompt style (teaching / interview / storytelling)

    Returns:
        A fully formatted prompt string.
    """
    # Step 1 — Validate inputs
    validated_audience, validated_tone = validate_inputs(audience, tone)

    # Step 2 — Select template
    if style not in style_templates:
        raise ValueError(f"Unknown style '{style}'. Choose from: {list(style_templates.keys())}")
    template = style_templates[style]

    # Step 3 — Format and return prompt
    return template.format(
        topic=topic,
        audience=validated_audience,
        tone=validated_tone
    )

# ── Test Outputs ───────────────────────────────────────────────────────────────
print("Task 6 — Prompt Generator App Output:")
print("-" * 60)

examples = [
    ("Neural Networks", "beginners", "fun",      "storytelling"),
    ("Python",          "expert",    "formal",    "interview"),
    ("Data Science",    "beginner",  "casual",    "teaching"),
]

for topic, audience, tone, style in examples:
    prompt = generate_prompt(topic, audience, tone, style)
    print(f"\n  style={style.upper()} | audience={audience} | tone={tone}")
    print(f"  → {prompt}")
````

## ✅ Task 7: Template Reusability Test (10%)

**Goal:** Use ONE template and run it with 5 different inputs.

Same structure → Different outputs = True reusability ✅


````python
# ── Task 7: Template Reusability Test ────────────────────────────────────────

# Single reusable template
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} to a {audience} in a {tone} and engaging way"
)

# 5 different inputs for the same template
reusability_inputs = [
    {"topic": "Artificial Intelligence", "audience": "school student",    "tone": "fun"},
    {"topic": "Cloud Computing",         "audience": "business executive","tone": "formal"},
    {"topic": "Blockchain",              "audience": "developer",         "tone": "technical"},
    {"topic": "Machine Learning",        "audience": "complete beginner", "tone": "simple"},
    {"topic": "Quantum Computing",       "audience": "researcher",        "tone": "detailed"},
]

print("Task 7 — Template Reusability Test (1 Template · 5 Inputs):")
print("=" * 65)
print("  Template: " + reusable_template.template)
print("=" * 65)

for i, inputs in enumerate(reusability_inputs, start=1):
    output = reusable_template.format(**inputs)
    print(f"\n  [{i}] topic={inputs['topic']} | audience={inputs['audience']} | tone={inputs['tone']}")
    print(f"       → {output}")

print("\n✅ Same template structure used across all 5 inputs — reusability confirmed!")
````

## 🔄 Full Pipeline Flow

```
User Input → Validation → Prompt Template → Dynamic Prompt Generation → Output
```

| Step | Component | Task |
|---|---|---|
| 1 | User Input | Provide topic, audience, tone, style |
| 2 | Validation | `validate_inputs()` — checks valid values / assigns defaults |
| 3 | Prompt Template | `PromptTemplate` / `ChatPromptTemplate` selected by style |
| 4 | Dynamic Generation | `template.format(...)` fills variables |
| 5 | Output | Final structured prompt ready for LLM |


````python
# ── End-to-End Pipeline Demo ──────────────────────────────────────────────────

print("🔄 End-to-End Pipeline Demo")
print("=" * 60)

user_topic    = "Deep Learning"
user_audience = "beginners"
user_tone     = "fun"
user_style    = "storytelling"

print(f"  Input  → topic: '{user_topic}' | audience: '{user_audience}' | tone: '{user_tone}' | style: '{user_style}'")

final_prompt = generate_prompt(
    topic    = user_topic,
    audience = user_audience,
    tone     = user_tone,
    style    = user_style
)

print(f"  Output → {final_prompt}")
print("\n✅ Pipeline complete — prompt is ready to be sent to an LLM!")
````
